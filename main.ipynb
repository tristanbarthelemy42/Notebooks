# La suite de Fibonacci

## Définition de la fonction fibonacci
En mathématiques, la suite de Fibonacci, tirant son nom du mathématicien italien Fibonacci, est une suite de nombres entiers dans laquelle chaque nombre est la somme des deux nombres qui le précèdent. Elle commence par les nombres 0 et 1[1] puis se poursuit avec :

1 (comme somme de 0 et 1),
2 (comme somme de 1 et 1),
3 (comme somme de 1 et 2),
5 (comme somme de 2 et 3),
8 (comme somme de 3 et 5),
etc.

Source : Wikipedia

```python
# Librairies Python requises
import matplotlib.pyplot as plt
```

```python
def fibonacci(n):
    """
    Retourne les n premiers termes de la suite de Fibonacci.

    La suite de Fibonacci est définie de la manière suivante :
        F(0) = 0
        F(1) = 1
        F(n) = F(n-1) + F(n-2) pour n >= 2

    Exemples :
        fibonacci(1) -> [0]
        fibonacci(2) -> [0, 1]
        fibonacci(6) -> [0, 1, 1, 2, 3, 5]

    Paramètre :
        n (int) : nombre de termes à générer.

    Valeur retournée :
        list : liste contenant les n premiers termes.
    """

    # Si n est inférieur ou égal à 0,
    # il n'y a aucun terme à retourner.
    if n <= 0:
        return []

    # Si on demande uniquement le premier terme,
    # on retourne [0].
    elif n == 1:
        return [0]

    # Si on demande les deux premiers termes,
    # on retourne [0, 1].
    elif n == 2:
        return [0, 1]

    # Pour n >= 3, on commence avec les deux
    # premiers termes déjà connus.
    fib_sequence = [0, 1]

    # La boucle commence à l'indice 2
    # (le troisième terme) et s'arrête à n - 1.
    for i in range(2, n):

        # Chaque nouveau terme est la somme
        # des deux termes précédents.
        next_value = fib_sequence[i - 1] + fib_sequence[i - 2]

        # On ajoute ce nouveau terme à la liste.
        fib_sequence.append(next_value)

    # On retourne la liste complète.
    return fib_sequence
```

```python
fib_sequence = fibonacci(10)

plt.plot(fib_sequence, marker='o')
plt.title("Suite de Fibonacci")
plt.xlabel("Index")
plt.ylabel("Valeur")
plt.grid()
plt.show()
```
